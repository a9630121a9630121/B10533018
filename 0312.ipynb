# 作業三 字串擷取處理

```python
import requests
from bs4 import BeautifulSoup
response = requests.get("http://www.novelscape.net/wxxs/j/jingyong/ffwz/001.htm")
response.encoding = "big5"
book1 = response.text
soup = BeautifulSoup(book1, 'html.parser')
sss=str(soup.find('hr'))
```

```python
sss[:]
```

```python
sss[sss.find('胡'):]
```

```python
sss[:sss.rfind('。')+1]
```

```python
sss[50:-612]
```

```python
sss[-612:50:-1]
```

```python
sss[:-612:100]
```

```python
sss[50::100]
```

```python
sss[::1000]
```
